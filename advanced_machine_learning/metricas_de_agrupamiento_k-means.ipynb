```python
# Paso 1: Importación de librerías necesarias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Ruta al archivo descargado
ruta_dataset = "../data/CC GENERAL.csv"  # Ajusta la ruta según tu ubicación

# Cargar el dataset
df = pd.read_csv(ruta_dataset)

# Mostrar las primeras filas
print("Primeras filas del dataset:")
print(df.head())

# Información general del dataset
print("\nInformación del dataset:")
print(df.info())

# Resumen estadístico
print("\nResumen estadístico:")
print(df.describe())

# Verificar valores nulos
print("\n¿Existen valores nulos?")
print(df.isnull().sum())
```

```python
# Verificar valores nulos
print("\n¿Existen valores nulos?")
print(df.isnull().sum())
```

```python
# Paso 2.1: Manejo de valores nulos
# Imputar valores nulos en 'CREDIT_LIMIT' con la mediana
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].median(), inplace=True)

# Imputar valores nulos en 'MINIMUM_PAYMENTS' con la mediana
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].median(), inplace=True)

# Verificar nuevamente los valores nulos
print("\n¿Existen valores nulos después de la imputación?")
print(df.isnull().sum())

# Paso 2.2: Selección de características relevantes
# Seleccionamos variables relacionadas con gastos, frecuencia y comportamiento
features = ['BALANCE', 'PURCHASES', 'CASH_ADVANCE', 'CREDIT_LIMIT', 'PAYMENTS']
X = df[features]

# Paso 2.3: Estandarización de las variables
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Verificar las variables estandarizadas
print("\nPrimeras filas de las variables estandarizadas:")
print(pd.DataFrame(X_scaled, columns=features).head())

```

```python
# Paso 3.1: Método del codo para K-means
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

sse = []
silhouette_scores = []
k_range = range(2, 11)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))

# Graficar el método del codo
plt.figure(figsize=(10, 6))
plt.plot(k_range, sse, marker='o', linestyle='--')
plt.title('Método del Codo para determinar K óptimo')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('SSE')
plt.xticks(k_range)
plt.grid()
plt.show()

# Graficar silhouette score
plt.figure(figsize=(10, 6))
plt.plot(k_range, silhouette_scores, marker='o', linestyle='--', color='orange')
plt.title('Silhouette Score para diferentes valores de K')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('Silhouette Score')
plt.xticks(k_range)
plt.grid()
plt.show()
```

```python
# Aplicar K-means con el número óptimo de clusters (ajusta según el método del codo)
k_optimo = 4  # Cambia este valor según el análisis
kmeans = KMeans(n_clusters=k_optimo, random_state=42)
kmeans_labels = kmeans.fit_predict(X_scaled)

# Visualizar los clusters
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=kmeans_labels, palette='viridis', s=100)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Centroides')
plt.title(f'Clusters formados por K-means con K={k_optimo}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.grid()
plt.show()
```

```python
from sklearn.cluster import DBSCAN

# Experimentar con diferentes valores de eps y min_samples
dbscan = DBSCAN(eps=1.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_scaled)

# Visualizar los clusters formados por DBSCAN
plt.figure(figsize=(10, 6))
sns.scatterplot(x=X_scaled[:, 0], y=X_scaled[:, 1], hue=dbscan_labels, palette='viridis', s=100)
plt.title('Clusters formados por DBSCAN')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid()
plt.show()
```

```python
from sklearn.metrics import silhouette_score, calinski_harabasz_score

# Verificar clusters detectados por DBSCAN
unique_labels = set(dbscan_labels)
print(f"Clusters detectados por DBSCAN: {unique_labels}")

# Calcular métricas si hay al menos 2 clusters válidos
n_clusters = len(unique_labels) - (1 if -1 in unique_labels else 0)

if n_clusters >= 2:
    # Calcular métricas para DBSCAN (excluyendo outliers)
    dbscan_silhouette = silhouette_score(X_scaled[dbscan_labels != -1], dbscan_labels[dbscan_labels != -1])
    dbscan_calinski = calinski_harabasz_score(X_scaled[dbscan_labels != -1], dbscan_labels[dbscan_labels != -1])

    print("\nMétricas para DBSCAN:")
    print(f"Silhouette Score: {dbscan_silhouette:.4f}")
    print(f"Calinski-Harabasz Index: {dbscan_calinski:.4f}")
else:
    print("\nDBSCAN no detectó suficientes clusters válidos para calcular métricas.")
```
