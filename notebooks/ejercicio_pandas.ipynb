Ejercicio pandas lectura de archivo csv

```python
import pandas as pd

# Ruta al archivo CSV
ruta_csv = r"C:\Users\59598\Documents\elidapena\retail-sales-analysis\data\vgsales.csv"

# Lee el archivo CSV en un DataFrame
df = pd.read_csv(ruta_csv)

# Mostrar las primeras 10 filas del DataFrame
print(df.head(10))
# Mostrar las últimas 5 filas del DataFrame
print(df.tail(5))
# Obtener información general del DataFrame
print(df.info())
# Generar estadísticas descriptivas
print(df.describe())
# Inspeccionar los tipos de datos de cada columna
print(df.dtypes)
# Contar los valores únicos en la columna Genre
print(df['Genre'].value_counts())
# Mostrar todos los valores únicos en la columna Platform
print(df['Platform'].unique())
# Filtrar las filas donde NA_Sales > 1 millón
ventas_na = df[df['NA_Sales'] > 1.0]
print(ventas_na)
# Filtrar las filas donde JP_Sales < 0.1 millón
ventas_jp = df[df['JP_Sales'] < 0.1]
print(ventas_jp)
# Filtrar las filas donde el género sea Action y Global_Sales > 2 millones
ventas_action = df.query("Genre == 'Action' and Global_Sales > 2.0")
print(ventas_action)
# Seleccionar y mostrar solo las columnas Name y Global_Sales
print(df[['Name', 'Global_Sales']])
# Seleccionar filas de la 5 a la 10 y las columnas Name y Genre
print(df.loc[5:10, ['Name', 'Genre']])
# Seleccionar las primeras 5 filas y las primeras 3 columnas
print(df.iloc[0:5, 0:3])




```
