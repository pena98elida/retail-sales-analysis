```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Ejemplo de datos
data = {
    'Calendar_Week': pd.date_range(start='2024-01-01', periods=52, freq='W'),
    'Sales': np.random.randint(1000, 5000, size=52)
}

df = pd.DataFrame(data)

# Configurar el tamao de la figura
plt.figure(figsize=(12, 6))

# Crear segmentacin de rangos de ventas
bins = [0, 2000, 3000, 4000, np.inf]  # Definir los rangos
labels = ['0-2000', '2000-3000', '3000-4000', '4000+']  # Etiquetas para los segmentos

df['Sales_Range'] = pd.cut(df['Sales'], bins=bins, labels=labels)

# Graficar por cada segmento
for label in labels:
    segment_data = df[df['Sales_Range'] == label]
    plt.plot(segment_data['Calendar_Week'], segment_data['Sales'], label=f'Ventas {label}')

# Ttulo y etiquetas
plt.title("Ventas a lo Largo del Tiempo con Segmentacin de Rango de Ventas", fontsize=16)
plt.xlabel("Fecha", fontsize=14)
plt.ylabel("Ventas", fontsize=14)

# Mejorar la visualizacin de las fechas en el eje x
plt.xticks(rotation=45)
plt.grid(True)

# Aadir la leyenda
plt.legend()

# Ajustar los mrgenes del grfico
plt.tight_layout()

# Mostrar el grfico
plt.show()
```
