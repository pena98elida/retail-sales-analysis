```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Ejemplo de datos
data = {
    'Calendar_Week': pd.date_range(start='2024-01-01', periods=52, freq='W'),
    'Sales': np.random.randint(1000, 5000, size=52)
}

df = pd.DataFrame(data)

# Configurar el tamaño de la figura
plt.figure(figsize=(12, 6))

# Crear segmentación de rangos de ventas
bins = [0, 2000, 3000, 4000, np.inf]  # Definir los rangos
labels = ['0-2000', '2000-3000', '3000-4000', '4000+']  # Etiquetas para los segmentos

df['Sales_Range'] = pd.cut(df['Sales'], bins=bins, labels=labels)

# Graficar por cada segmento
for label in labels:
    segment_data = df[df['Sales_Range'] == label]
    plt.plot(segment_data['Calendar_Week'], segment_data['Sales'], label=f'Ventas {label}')

# Título y etiquetas
plt.title("Ventas a lo Largo del Tiempo con Segmentación de Rango de Ventas", fontsize=16)
plt.xlabel("Fecha", fontsize=14)
plt.ylabel("Ventas", fontsize=14)

# Mejorar la visualización de las fechas en el eje x
plt.xticks(rotation=45)
plt.grid(True)

# Añadir la leyenda
plt.legend()

# Ajustar los márgenes del gráfico
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```
